# Q1) Is their any relation between number of double faults commited by men and women

```python
m1=(aus_men['DBF.1']+aus_men['DBF.2']).mean()
m2=(french_men['DBF.1']+french_men['DBF.2']).mean()
m3=(US_men['DBF.1']+US_men['DBF.1']).mean()
m4=(Wimbledon_men['DBF.1']+Wimbledon_men['DBF.2']).mean()
mean1=(m1+m2+m3+m4).mean()

w1=(aus_women['DBF.1']+aus_men['DBF.2']).mean()
w2=(french_women['DBF.1']+french_men['DBF.2']).mean()
w3=(US_women['DBF.1']+US_men['DBF.1']).mean()
w4=(Wimbledon_women['DBF.1']+Wimbledon_men['DBF.2']).mean()
mean2=(w1+w2+w3+w4).mean()

dic={'DBF':[m1,m2,m3,m4,w1,w2,w3,w4]
    ,'Tournament':['Aus open','french open','US open','Wimbledon open']*2
    ,'gender':['men']*4 + ['women']*4}
df=pd.DataFrame(dic)

sns.barplot(data=df,x='Tournament',y='DBF',hue='gender')
plt.figure()
plt.pie([mean1,mean2],labels=['Men','Women'],wedgeprops={'edgecolor':'Black','linewidth':3},autopct='%.2f %%')
plt.show()
```
