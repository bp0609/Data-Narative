### Adding gender and tournament column in each dataframe

```python
def columnadder(df,gender,tournament):
    df['Gender'],df['Tournament']=pd.Series([gender]*df.shape[0]),pd.Series([tournament]*df.shape[0])

columnadder(aus_men,"Men",'AusOpen')
columnadder(french_men,"Men",'FrenchOpen')
columnadder(US_men,"Men",'USOpen')
columnadder(Wimbledon_men,"Men",'WimbledonOpen')
columnadder(aus_women,"Women",'AusOpen')
columnadder(french_women,"Women",'FrenchOpen')
columnadder(US_women,"Women",'USOpen')
columnadder(Wimbledon_women,"Women",'WimbledonOpen')
aus_men.head()
```

### Q1) Is their any relation between number of double faults commited by men and women

```python
m1=(aus_men['DBF.1']+aus_men['DBF.2']).sum()
m2=(french_men['DBF.1']+french_men['DBF.2']).sum()
m3=(US_men['DBF.1']+US_men['DBF.1']).sum()
m4=(Wimbledon_men['DBF.1']+Wimbledon_men['DBF.2']).sum()
mean1=(m1+m2+m3+m4).mean()

w1=(aus_women['DBF.1']+aus_men['DBF.2']).sum()
w2=(french_women['DBF.1']+french_men['DBF.2']).sum()
w3=(US_women['DBF.1']+US_men['DBF.1']).sum()
w4=(Wimbledon_women['DBF.1']+Wimbledon_men['DBF.2']).sum()
mean2=(w1+w2+w3+w4).mean()

dic={'No. of Double Faults':[m1,m2,m3,m4,w1,w2,w3,w4]
    ,'Tournament':['Aus open','french open','US open','Wimbledon open']*2
    ,'gender':['men']*4 + ['women']*4}
df=pd.DataFrame(dic)

sns.barplot(data=df,x='Tournament',y='No. of Double Faults',hue='gender')
plt.figure()

plt.pie([mean1,mean2],labels=['Men','Women'],wedgeprops={'edgecolor':'Black','linewidth':3},autopct='%.2f %%',shadow=True)
plt.title('Overall percentage of Double Faults committed by ')
plt.show()
df
```

### Q) What can you say about number of double faults and number of aces, increases or decreases as the tournament proceeds to furthur rounds 

```python
# round 1 to 7
r1= df_merge[df_merge['Round']==1]
r2= df_merge[df_merge['Round']==2]
r3= df_merge[df_merge['Round']==3]
r4= df_merge[df_merge['Round']==4]
r5= df_merge[df_merge['Round']==5]
r6= df_merge[df_merge['Round']==6]
r7= df_merge[df_merge['Round']==7]

lst=[r1,r2,r3,r4,r5,r6,r7]

df_lst=[]
for df in lst:
    s1=(df['DBF.1'] + df['DBF.2']).mean()
    s2=(df['ACE.1'] + df['ACE.2']).mean()
    df_lst=df_lst+[s1.round(),s2.round()]  

dic={'Round':['Round 1','Round 1','Round 2','Round 2','Round 3','Round 3','Round 4','Round 4','Round 5','Round 5','Round 6','Round 6','Round 7','Round 7']
     ,'No. of faults and aces':df_lst
     ,'Type':['Double Faults','Aces']*7}

df=pd.DataFrame(dic)
sns.barplot(data=df,x='Round',y='No. of faults and aces',hue='Type',palette=["#FF66B2","#FFB266"])
Aces=[]
DBF=[]
for i in range(1,14,2):
    Aces=Aces+[df_lst[i]]
for i in range(0,14,2):
    DBF=DBF+[df_lst[i]]
plt.plot(range(0,7),Aces,color='#FF8000',linewidth=3,marker='o',linestyle='-',markersize=10)
plt.plot(range(0,7),DBF,color='#FF007F',linewidth=3,marker='o',linestyle='-',markersize=10)
# plt.plot(range(1,11),s,color='red',linewidth=3,marker='o',linestyle='-',markersize=10)
plt.figure()    
lst=[]
for i in range(7):
    lst=lst+[(df_lst[2*i+1]/df_lst[2*i])]

plt.plot(range(0,7),lst,color='#FF00FF',linewidth=3,marker='o',linestyle='-',markersize=5)
plt.xticks(ticks=range(7),labels=['Round 1','Round 2','Round 3','Round 4','Round 5','Round 6','Round 7'])
plt.xlabel('Rounds')
plt.ylabel('Ratio of Aces to double faults')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
# from sklearn.metrics import accuracy_score
model=LinearRegression(fit_intercept=True)
df_=df_merge.loc[:,['FSP.1','ACE.1','DBF.1','BPW.1','Result']].dropna()
test_matrix=np.array(df_.loc[:,['FSP.1','ACE.1','DBF.1','BPW.1']])
test_result=np.array(df_.loc[:,'Result'])
lst=[aus_men,aus_women,french_men,french_women,US_men,US_women,Wimbledon_men,Wimbledon_women]

model_accuracy=[]
for i in lst:
    df=i.loc[:,['FSP.1','ACE.1','DBF.1','BPW.1','Result']].dropna()
    feature_matrix=np.array(df.loc[:,['FSP.1','ACE.1','DBF.1','BPW.1']])
    target_matrix=np.array(df.loc[:,'Result'])
    #Training the Model
    model.fit(feature_matrix,target_matrix)
    pred_result= model.predict(test_matrix)
    error=mean_squared_error(test_result, pred_result)
    model_accuracy.append(error)

plt.plot(range(8),model_accuracy,color='#FF8000',linewidth=3,marker='o',linestyle='-',markersize=10)
plt.xticks(ticks=range(8),labels=['aus_m','aus_w','fre_m','fre_w','US_m','US_w','Wim_m','Wim_w'],)
plt.xlabel('Model trained from different datasets')
plt.ylabel('Error of different models')
plt.show()

model.intercept_
```

```python
df=df_merge
def prob_losing(set_1,set_2,df):
    df1=df[df['Result']==0]
    df1_=df1[df1[set_1]<df1[set_2]]
    df2=df[df['Result']==1]
    df2_=df2[df2[set_1]>df2[set_2]]
    prob_set=(len(df2_)+len(df1_))/(len(df1)+len(df2))
    return prob_set

probability1=prob_losing('ST1.1','ST1.2',df)
probability2=prob_losing('ST2.1','ST2.2',df)
probability3=prob_losing('ST3.1','ST3.2',df)


plt.plot(range(1,4),[probability1,probability2,probability3],color='red',linewidth=3,marker='o',linestyle='-',markersize=10)

bars=plt.bar(np.arange(1,4),[probability1,probability2,probability3],color='#9966FF')
plt.xticks(ticks=np.arange(1,4),labels=['Set 1 lose','Set 2 lose','Set 3 lose'])
lst=list('/o*')
for i in range(3):
  bars[i].set_hatch(lst[i])
plt.ylabel('Probability of losing the match',fontsize=15)
plt.xlabel('Sets',fontsize=15)
plt.show()
```
